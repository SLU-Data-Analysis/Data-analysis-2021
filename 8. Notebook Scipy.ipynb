# SciPy

SciPy v Pythonu je open-source knihovna používaná k řešení matematických, vědeckých a technických problémů. Umožňuje uživatelům manipulovat s daty a vizualizovat je pomocí široké škály příkazů Pythonu. SciPy je postaven na Python knihovně NumPy.

## Proč používat SciPy

* SciPy obsahuje různé dílčí balíčky, které pomáhají vyřešit nejběžnější problémy související s vědeckými výpočty.
* Balíček SciPy v Pythonu je nejpoužívanější vědeckou knihovnou vyjma GNU Scientific Library pro C/C++ nebo Matlab.
* Snadné použití a pochopení stejně jako rychlý výpočetní výkon.
* Funguje v kombinaci s Numpy poli.

## Numpy vs. SciPy

Numpy:
* Numpy je napsáno v C a používá se pro matematické nebo numerické výpočty.
* Je rychlejší než jiné Python knihovny.
* Numpy je nejužitečnější knihovnou pro datovou analýzu na základní matematické nebo numerické výpočty.
* Numpy neobsahuje nic jiného než datový typ pole, který podporuje nejzákladnější operace, jako je seřazení, shaping, indexování atd.

SciPy:
* SciPy je postaven na knihovně NumPy.
* Modul SciPy je plně vybavenou verzí lineární algebry, zatímco Numpy obsahuje pouze několik funkcí pro lineární algebru.
* Většina nových funkcí pro datovou analýzu jsou k dispozici spíše v SciPy než Numpy.

## Vlastní čísla a vlastní vektory

* Nejběžnějším problémem lineární algebry jsou vlastní hodnoty a vlastní vektory, které lze snadno vyřešit pomocí funkce eig().
* Nyní najdeme vlastní hodnotu (X) a odpovídající vlastní vektor dvourozměrné čtvercové matice.

````python
from scipy import linalg
import numpy as np
#Vytvoříme dvourozměrnou matici
arr = np.array([[5,4],[6,3]])
#Předáme hodnoty do naší funkce
eg_val, eg_vect = linalg.eig(arr)
#Zobrazíme vlastní čísla
print(eg_val)
#Zobrazíme vlastní vektory
print(eg_vect)
````

## Diskrétní Fourierova transformace

* Diskrétní Fourierova transformace je matematická technika, která se používá při převodu prostorových dat na frekvenční data.
* FFT (Fast Fourier Transformation) je algoritmus pro výpočet Diskrétní Fourierovy transformace.
* FFT se aplikuje na vícerozměrné pole.
* Frekvence definuje počet signálů nebo vlnových délek v konkrétním časovém období.

````python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np 

#Frekvence v hz
fre  = 5 
#Vzorkovací frekvence
fre_samp = 50
t = np.linspace(0, 2, 2 * fre_samp, endpoint = False )
a = np.sin(fre  * 2 * np.pi * t)
figure, axis = plt.subplots()
axis.plot(t, a)
axis.set_xlabel ('Time (s)')
axis.set_ylabel ('Signal amplitude')
plt.show()
````

Frekvence je 5 Hz a její signál se opakuje za 1/5 sekundy. .

````python
from scipy import fftpack

A = fftpack.fft(a)
frequency = fftpack.fftfreq(len(a)) * fre_samp
figure, axis = plt.subplots()

axis.stem(frequency, np.abs(A), use_line_collection=True)
axis.set_xlabel('Frequency in Hz')
axis.set_ylabel('Frequency Spectrum Magnitude')
axis.set_xlim(-fre_samp / 2, fre_samp/ 2)
axis.set_ylim(-5, 110)
plt.show()
````

* Můžeme si všimnout, že výstup je jednorozměrné pole.
* Vstup obsahující komplexní hodnoty jsou nuly kromě dvou bodů.
* V příkladu DFT vizualizujeme velikost signálu.

## Integrace pomocí SciPy – Numerické integrace

* Když integrujeme jakoukoli funkci, kde analytická integrace není možná, musíme se obrátit na numerickou integraci.
* SciPy poskytuje funkce pro numerickou integraci matematických funkcí.
* Knihovna scipy.integrate má pravidla pro jednoduchou integraci, dvojitou integraci, trojitou integraci, vícenásobnou integraci, Gaussův kvadrát, Rombergovo a Simpsonovo pravidlo.

````python
# Jednoduchá integrace
from scipy import integrate
# Funkce f(x) jako f
f = lambda x : x**2
#Jedna iterace pro a = 0 & b = 1  
integration = integrate.quad(f, 0 , 1)
print(integration)
````

Výstupem funkce jsou dvě hodnoty, ve kterých první hodnotou je integrace a druhou hodnotou je odhadovaná chyba v integrálu.

````python
# Dvojitá integrace
from scipy import integrate
import numpy as np
#importujeme funkci pro odmocninu z knihovny math
from math import sqrt
# Definujeme funkci f(x)
f = lambda x, y : 64 *x*y
# Spodní limita druhého integrálu
p = lambda x : 0
# Vrchní limita prvního integrálu
q = lambda y : sqrt(1 - 2*y**2)
# Provedeme dvojitou integraci
integration = integrate.dblquad(f , 0 , 2/4,  p, q)
print(integration)
````

Znovu máme jako výstup dvě hodnoty, kde první je integrál a druhá je odhadnutá chyba v integrálu.
