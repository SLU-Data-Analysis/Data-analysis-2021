## Spojnicový graf

Spojnicový graf je jednou z nejjednodušších a nejpoužívanějších technik vizualizace dat. Spojnicový graf zobrazuje informace jako řadu datových bodů nebo značek spojených přímkami. Můžete upravit tvar, velikost, barvu a další estetické prvky čar a značek pro lepší přehlednost.

Zde je Python seznam, který nám ukazuje výnos jablek (tun na hektar) po dobu šesti let v imaginární zemi zvané Kanto.

````python
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
````

Pomocí spojnicového grafu můžeme vykreslit, jak se výnos jablek mění. K tomu můžeme použít funkci `plt.plot`.

````python
plt.plot(yield_apples)
````

Vyvoláním funkce `plt.plot` se nakreslí spojnicový graf. Ve výstupu se také zobrazí seznam vykreslených grafů „[<matplotlib.lines.Line2D at 0x7ff70aa20760>]“. Na konec posledního příkazu v buňce můžeme přidat středník (`;`), abychom se tomuto zobrazení vyhnuli. Takto se nám vykreslí pouze graf.

````python
plt.plot(yield_apples);
````

Pojďme spolu tento graf postupně vylepšit, abychom se v něm lépe vyznali.

````python
years = [2010, 2011, 2012, 2013, 2014, 2015]
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
````

````python
plt.plot(years, yield_apples);
````

### Názvy os

Pomocí metod `plt.xlabel` a` plt.ylabel` můžeme přidat popisky k osám, abychom ukázali, co každá osa představuje.

### Vykreslení více spojnicových grafů

Funkci `plt.plot` můžete vyvolat jednou pro každý spojnicový graf a vykreslit více čar do stejného grafu. Porovnejme výnosy jablek vs. pomerančů v Kantu.

````python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
````

````python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
````

### Titulek grafu a legenda

Abychom rozlišili mezi vykreslenými spojnicovými grafy, můžeme do grafu zahrnout legendu pomocí funkce `plt.legend`. Můžeme také nastavit název grafu pomocí funkce `plt.title`.

````python
plt.plot(years, apples)
plt.plot(years, oranges)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
````

### Markery

Pomocí argumentu `marker` v` plt.plot` můžeme také vytvořit markery pro datové body na každém grafu. Matplotlib poskytuje mnoho různých markerů, jako je kruh, kříž, čtverec, diamant atd. Celý seznam typů markerů najdete zde: https://matplotlib.org/3.1.1/api/markers_api.html.

````python
plt.plot(years, apples, marker='o')
plt.plot(years, oranges, marker='x')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
````

### Styly čar a markerů

Funkce `plt.plot` podporuje mnoho argumentů pro změnu stylů čáry a markeru:

- `color` nebo `c`: Změní barvu čáry ([podporované barvy](https://matplotlib.org/3.1.0/gallery/color/named_colors.html))
- `linestyle` nebo `ls`: Vybere mezi plnou a přerušovanou čarou
- `linewidth` nebo `lw`: Nastaví šířku čárky
- `markersize` nebo `ms`: Nastaví velikosti markerů
- `markeredgecolor` nebo `mec`: Nastaví okrajovou barvu markerů
- `markeredgewidth` nebo `mew`: Nastaví šířku okrajů markerů
- `markerfacecolor` nebo `mfc`: Nastaví výplňovou barvu markerů
- `alpha`: Průhlednost grafu

````python
plt.plot(years, apples, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, oranges, marker='o', c='r', ls='--', lw=3, ms=10, alpha=.5)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
````

Jako zkratku pro specifikování tvaru markeru a stylu a barvy čáry můžeme použít argument `fmt`, který bude třetí v pořadí funkce `plt.plot`.
```
fmt = '[marker][line][color]'
```

````python
plt.plot(years, apples, 's-b')
plt.plot(years, oranges, 'o--r')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
````

## Bodový diagram

V bodovém grafu jsou hodnoty 2 proměnných vyneseny jako body na 2rozměrné mřížce. Dále můžete použít třetí proměnnou k určení velikosti nebo barvy bodů. Vyzkoušejme si to na příkladu.

Dataset [Iris flower](https://en.wikipedia.org/wiki/Iris_flower_data_set) poskytuje ukázková měření kalíšků a lístků pro tři druhy květin. Datová sada Iris je součástí knihovny Seaborn a lze ji načíst jako datový rámec Pandas.

````python
flowers_df = sns.load_dataset("iris")
````

````python
flowers_df
````

````python
flowers_df.species.unique()
````

Zkusme si vykreslit vztah mezi délkou kalíšku a mezi jeho šířkou. Naším prvním instinktem může být vytvoření spojnicového grafu pomocí `plt.plot`.

````python
plt.plot(flowers_df.sepal_length, flowers_df.sepal_width);
````

Jak vidíme, výstup nám toho moc neřekne, protože v datové sadě je příliš mnoho kombinací těchto dvou vlastností. Zdá se, že mezi nimi není jednoduchý vztah.

Můžeme použít bodový graf k vizualizaci toho, jak se liší délka a šířka kalíšku pomocí funkce `scatterplot` z modulu` seaborn` (importováno jako `sns`). 

````python
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width);
````

### Přidání rozlišovacích značek „Hues“

Všimněte si, že body ve výše uvedeném grafu tvoří odlišné shluky s některými odlehlými hodnotami. Druhy květů můžeme vybarvit pomocí rozlišovacích značek (tzv. „hues“). Body můžeme také zvětšit pomocí argumentu `s`.

````python
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width, hue=flowers_df.species, s=100);
````

Přidáním barevných značek se graf stává mnohem přehlednějším. Okamžitě si můžeme všimnout, že květiny Setosa mají menší délku kalíšku, ale vyšší šířku kališku. Oproti tomu u květin Virginica je to naopak.

### Přizpůsobení Seaborn grafů

Vzhledem k tomu, že knihovna Seaborn používá interně vykreslovací funkce Matplotlib, můžeme k úpravě grafů použít funkce jako `plt.figure` a` plt.title`.

````python
plt.figure(figsize=(12, 6))
plt.title('Sepal Dimensions')

sns.scatterplot(x=flowers_df.sepal_length, 
                y=flowers_df.sepal_width, 
                hue=flowers_df.species,
                s=100);
````

## Histogram

Histogram představuje distribuci proměnné vytvořením binů (intervalů) v rozsahu hodnot a zobrazením svislých pruhů označujících počet pozorování v každém binu.

Podívejme se například na distribuci hodnot šířky kalíšku v datasetu flowers. K vytvoření histogramu můžeme použít funkci `plt.hist`.

````python
flowers_df = sns.load_dataset("iris")
````

````python
flowers_df.sepal_width
````

````python
plt.title("Distribution of Sepal Width")
plt.hist(flowers_df.sepal_width);
````

Hned si můžeme všimnout, že šířky kalíšku leží v rozmezí 2,0 – 4,5 a kolem 35 hodnot je v rozmezí 2,9 – 3,1, což se jeví jako největší bin.

### Ovládání velikosti a počtu binů

Můžeme si přizpůsobit počet binů nebo velikost každého z nich použitím argumentů pro biny.

````python
plt.hist(flowers_df.sepal_width, bins=5);
````

````python
import numpy as np

# Určení hranic každého binu
plt.hist(flowers_df.sepal_width, bins=np.arange(2, 5, 0.25));
````

````python
# Biny nestejné velikosti
plt.hist(flowers_df.sepal_width, bins=[1, 3, 4, 4.5]);
````

### Více Histogramů

Podobně jako spojnicové grafy můžeme nakreslit více histogramů do jednoho grafu. Můžeme snížit neprůhlednost každého histogramu tak, aby jeden pruh histogramu neskrýval ostatní.

Nakreslíme samostatné histogramy pro každý druh květin.

````python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
````

````python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
````

Můžeme také skládat více histogramů na sebe.

````python
plt.title('Distribution of Sepal Width')

plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True);

plt.legend(['Setosa', 'Versicolor', 'Virginica']);
````

## Sloupcový graf

Sloupcové grafy jsou velmi podobné spojnicovým grafům, tj. ukazují posloupnost hodnot. U každé hodnoty je však zobrazen pruh, nikoli body spojené čarami. K nakreslení sloupcového grafu můžeme použít funkci `plt.bar`.

````python
years = range(2000, 2006)
apples = [0.35, 0.6, 0.9, 0.8, 0.65, 0.8]
oranges = [0.4, 0.8, 0.9, 0.7, 0.6, 0.8]
````

````python
plt.bar(years, oranges);
````

Stejně jako histogramy můžeme skládat sloupce na sebe. Abychom toho dosáhli, použijeme argument `bottom` funkce` plt.bar`.

````python
plt.bar(years, apples)
plt.bar(years, oranges, bottom=apples);
````

## Boxplot grafy

Boxplot je metoda pro grafické znázornění skupin číselných dat prostřednictvím jejich kvartilů. Boxploty mohou mít také čáry vyčnívající z krabiček (vousy), které označují variabilitu mimo horní a dolní kvartily. Outliers mohou být vyneseny jako jednotlivé body.

````python
import seaborn as sns

sns.set_style("whitegrid")

tips = sns.load_dataset("tips")
````

````python
ax = sns.boxplot(x="day", y="total_bill", data=tips)
````

Boxplot je standardizovaný způsob zobrazení datové sady na základě souhrnu pěti čísel: minima, maxima, mediánu vzorku a prvního a třetího kvartilu.

- **Minimum** (Q<sub>0</sub> nebo nultý percentil): nejnižší datový bod bez jakýchkoli odlehlých hodnot.
- **Maximum** (Q<sub>4</sub> nebo 100. percentil): nejvyšší datový bod bez jakýchkoli odlehlých hodnot.
- **Medián** (Q<sub>2</sub> nebo 50. percentil): střední hodnota datové sady.
- **První kvartil** (Q<sub>1</sub> nebo 25. percentil): také známý jako dolní kvartil (0,25), je mediánem dolní poloviny datové sady.
- **Třetí kvartil** (Q<sub>3</sub> nebo 75. percentil): také známý jako horní kvartil (0,75), je mediánem horní poloviny datové sady.

Důležitým prvkem použitým ke konstrukci boxplotu, který ale není součástí výše uvedeného pětimístného souhrnu, je mezikvartilní rozsah nebo IQR označený níže:

- **Mezikvartilní rozsah (IQR)**: je vzdálenost mezi horním a dolním kvartilem.

$ IQR=Q_3-Q_1 = q_n(0.75)-q_n(0.25) $

Boxplot je sestaven ze dvou částí, boxu a sady vousů znázorněných na grafu výše. Nejnižší bod je minimum datové sady a nejvyšší bod je maximum datové sady. Pole je nakresleno od Q1 do Q3 s vodorovnou čarou nakreslenou uprostřed, která označuje medián.

````python
ax = sns.boxplot(x="day", y="total_bill", hue="smoker",

                 data=tips, palette="Set3")
````

## Teplotní mapa

Teplotní mapa se používá k vizualizaci dvourozměrných dat, jako je matice nebo tabulka, pomocí barev. Nejlépe tomu porozumíme na příkladu. Použijeme další ukázkovou datovou sadu od Seabornu zvanou `flights` a její pomocí si vizualizujeme počet cestujících každý měsíc na letišti po dobu 12 let.

````python
flights_df = sns.load_dataset("flights").pivot("month", "year", "passengers")
````

````python
flights_df
````

`flight_df` je matice s jedním řádkem pro každý měsíc a jedním sloupcem pro každý rok. Hodnoty ukazují počet cestujících (v tisících), kteří navštívili letiště v konkrétním měsíci roku. Můžeme použít funkci `sns.heatmap` k vizualizaci počtu lidí na letišti.

````python
plt.title("No. of Passengers (1000s)")
sns.heatmap(flights_df);
````

Jasnější barvy naznačují vyšší počet lidí na letišti. Při pohledu na graf můžeme odvodit dvě věci:

- Počty lidí na letišti v daném roce bývají nejvyšší kolem července a srpna.
- Počty lidí na letišti v jakémkoliv měsící rostou s přibývajícím rokem.

Skutečné hodnoty v každém bloku můžeme také zobrazit zadáním hodnoty `annot = True` a použitím argumentu` cmap` můžeme změnit palety barev.

````python
plt.title("No. of Passengers (1000s)")
sns.heatmap(flights_df, fmt="d", annot=True, cmap='Blues');
````

## Vykreslení více grafů do mřížky

Knihovny Matplotlib a Seaborn také podporují vykreslování více grafů do mřížky pomocí příkazu `plt.subplots`, který vrací sadu os pro vykreslení.

````python
fig, axes = plt.subplots(2, 2, figsize=(16, 8))

# Use the axes for plotting
axes[0,0].plot(years, apples, 's-b')
axes[0,0].plot(years, oranges, 'o--r')
axes[0,0].set_xlabel('Year')
axes[0,0].set_ylabel('Yield (tons per hectare)')
axes[0,0].legend(['Apples', 'Oranges']);
axes[0,0].set_title('Crop Yields in Kanto')


# Pass the axes into seaborn
axes[0,1].set_title('Sepal Length vs. Sepal Width')
sns.scatterplot(x=flowers_df.sepal_length, 
                y=flowers_df.sepal_width, 
                hue=flowers_df.species, 
                s=100, 
                ax=axes[0,1]);

# Pass the axes into seaborn
axes[1,0].set_title('Restaurant bills')
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df, ax=axes[1,0]);

# Pass the axes into seaborn
axes[1,1].set_title('Flight traffic')
sns.heatmap(flights_df, cmap='Blues', ax=axes[1,1]);

plt.tight_layout(pad=2);
````
